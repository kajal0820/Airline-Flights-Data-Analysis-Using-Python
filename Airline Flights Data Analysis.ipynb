#### showing all the airlines with there frequencies

#### showing all the airlines with there Number of Flights In Horizontal Bar Graph

```python
data['airline'].value_counts(ascending=True).plot.barh(color = ['lightgreen', 'lightblue'])
```

```python
plt.title("Airlines with frequencies")
plt.xlabel(" Number of flights")
plt.ylabel(" Airlines")
plt.show()
```

#### Showing the departure time for the flight

```python
data['departure_time'].value_counts()
```

#### Showing the departure time and the arrival time of flights with there counts

```python
plt.figure(figsize = (16,4))

plt.subplot(1,2,1)

plt.bar( data['departure_time'].value_counts().index, data['departure_time'].value_counts().values, color=['r','b'])

plt.title("Departure Time")
plt.xlabel("D. Time")
plt.ylabel("D. Freq")

plt.subplot(1,2,2)

plt.bar(data['arrival_time'].value_counts().index, data['arrival_time'].value_counts().values, color=['g', 'y'])

plt.title("Arrival Time")

plt.xlabel("A. Time") 
plt.ylabel("A. Fres")

plt.show()
```

#### Showing the Source of the flights

```python
data['source_city'].value_counts()
```

#### Showing the Source city and the destination city for the flights with there counts

```python
plt.figure(figsize = (16,4))

plt.subplot(1,2,1)

plt.bar( data['source_city'].value_counts().index, data['source_city'].value_counts().values, color=['m','g'])

plt.title("Source Cities with Number Of Flights")
plt.xlabel("Cities")
plt.ylabel("Number of Flights")

plt.subplot(1,2,2)

plt.bar(data['destination_city'].value_counts().index, data['destination_city'].value_counts().values, color=['r', 'y'])

plt.title("Destination Cities  with number of Flights")

plt.xlabel("A. Cities") 
plt.ylabel("A. No.of Flights")

plt.show()
```
